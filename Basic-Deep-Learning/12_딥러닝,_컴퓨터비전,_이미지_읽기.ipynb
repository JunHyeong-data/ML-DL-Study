```python
import torch
from torchvision import datasets, transforms

MNIST_trainset = datasets.MNIST(root='./data', train=True, download=True)
print(len(MNIST_trainset))
MNIST_img, MNIST_label = MNIST_trainset[0]
print(MNIST_trainset[2])
```

```python
import matplotlib.pyplot as plt
import numpy as np

MNIST_img_array = np.array(MNIST_img)
print(MNIST_img_array.shape)
plt.imshow(MNIST_img_array, cmap='gray')
plt.title(f'Label: {MNIST_label}')
plt.show()
```

```python
np.set_printoptions()
np.set_printoptions(linewidth=160)
print(MNIST_img_array)
#np.set_printoptions(): NumPy가 데이터를 화면에 출력하는 방식을 결정하는 설정 함수입니다. 출력되는 소수점 자리수, 생략 여부, 한 줄의 길이 등을 조절할 수 있습니다.
```
